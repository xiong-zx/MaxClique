```python
def delete_graphs(graph_list, indices_list):
  new_list = []
  for i in range(len(graph_list)):
    if i not in indices_list:
      new_list.append(graph_list[i])
  return np.array(new_list)
```

```python
# densities of sthe 100 graphs
graph_densities= np.load('/content/drive/MyDrive/Colab Notebooks/Decomp/Results/densities.npy')

# Calculate the average runtime for each algorithm within a density range
def average_runtime_by_density(algorithm_iteration, densities, density_window):
    averages = []
    for start_density in np.arange(0.1, 0.7, density_window):
        end_density = start_density + density_window
        mask = np.logical_and(densities >= start_density, densities < end_density)
        subset_runtimes = algorithm_iteration[mask]
        if len(subset_runtimes) > 0:
            averages.append(np.mean(subset_runtimes))
        else:
            averages.append(0)  # Handle the case where there are no data points in the density range
    return averages
```

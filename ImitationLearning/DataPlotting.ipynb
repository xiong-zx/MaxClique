```python
iterations = [bc_iteration, simple_iteration, pairwise_iteration, smart_iteration, dbk_iteration, max_iteration, random_iteration]
subgraphs = [bc_subgraph, simple_subgraph, pairwise_subgraph, smart_subgraph, dbk_subgraph, max_subgraph, random_subgraph]
times = [bc_time, simple_time, pairwise_time, smart_time, dbk_time, max_time, random_time]
```

```python
normalized_iterations = []
for i in range(iterations[0].size):
    elements = [array[i] for array in iterations]
    min_val = min(elements)
    max_val = max(elements)
    normalized_element = [0.0]*len(subgraphs).size if min_val == max_val else [(element - min_val) / (max_val - min_val) for element in elements]
    normalized_iterations.append(normalized_element)
normalized_iterations = np.array(normalized_iterations).T.tolist()

normalized_subgraphs = []
for i in range(subgraphs[0].size):
    elements = [array[i] for array in subgraphs]
    min_val = min(elements)
    max_val = max(elements)
    normalized_element = [0.0]*len(subgraphs) if min_val == max_val else [(element - min_val) / (max_val - min_val) for element in elements]
    normalized_subgraphs.append(normalized_element)
normalized_subgraphs = np.array(normalized_subgraphs).T.tolist()

normalized_times = []
for i in range(times[0].size):
    elements = [array[i] for array in times]
    min_val = min(elements)
    max_val = max(elements)
    normalized_element = [0.0]*len(subgraphs).size if min_val == max_val else [(element - min_val) / (max_val - min_val) for element in elements]
    normalized_times.append(normalized_element)
normalized_times = np.array(normalized_times).T.tolist()
```

```python
# Create a boxplot
fig, ax = plt.subplots(figsize=(10, 6))
plt.boxplot(normalized_iterations, labels=['Binary Class', 'SimpleRank', 'PairwiseRank*', "Smart", "Min Degree (DBK)", "Max Degree", "Random"])
plt.title('Boxplot of Number of Iterations')
plt.ylabel('Iterations')
# Show the plot
plt.show()

fig, ax = plt.subplots(figsize=(10, 6))
plt.boxplot(normalized_subgraphs, labels=['Binary Class', 'SimpleRank', 'PairwiseRank*', "Smart", "Min Degree (DBK)", "Max Degree", "Random"])
plt.title('Boxplot of Number of Subgraphs')
plt.ylabel('Subgraphs')
# Show the plot
plt.show()

fig, ax = plt.subplots(figsize=(10, 6))
plt.boxplot(normalized_times, labels=['Binary Class', 'SimpleRank', 'PairwiseRank*', "Smart", "Min Degree (DBK)", "Max Degree", "Random"])
plt.title('Boxplot of Time Spent')
plt.ylabel('Time')
# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# (times, labels=['Binary Class', 'SimpleRank', 'PairwiseRank*', "Smart", "Min Degree (DBK)", "Max Degree", "Random"])
# Create dictionaries to store runtimes for each graph size
bc = {size: runtime for size, runtime in zip(graph_sizes, iterations[0])}
simple = {size: runtime for size, runtime in zip(graph_sizes, iterations[1])}
pairwise = {size: runtime for size, runtime in zip(graph_sizes, iterations[2])}
smart = {size: runtime for size, runtime in zip(graph_sizes, iterations[3])}
min_degree = {size: runtime for size, runtime in zip(graph_sizes, iterations[4])}
max_degree = {size: runtime for size, runtime in zip(graph_sizes, iterations[5])}
random = {size: runtime for size, runtime in zip(graph_sizes, iterations[6])}

# Calculate average runtimes and standard deviations for each graph size
average_bc = [np.mean(bc[size]) for size in sorted(bc.keys())]
std_bc = [np.std(bc[size]) for size in sorted(bc.keys())]

average_simple = [np.mean(simple[size]) for size in sorted(simple.keys())]
std_simple = [np.std(simple[size]) for size in sorted(simple.keys())]

average_pairwise = [np.mean(pairwise[size]) for size in sorted(pairwise.keys())]
std_pairwise = [np.std(pairwise[size]) for size in sorted(pairwise.keys())]

average_smart = [np.mean(smart[size]) for size in sorted(smart.keys())]
std_smart = [np.std(smart[size]) for size in sorted(smart.keys())]

average_min_degree = [np.mean(min_degree[size]) for size in sorted(min_degree.keys())]
std_min_degree = [np.std(min_degree[size]) for size in sorted(min_degree.keys())]

average_max_degree = [np.mean(max_degree[size]) for size in sorted(max_degree.keys())]
std_max_degree = [np.std(max_degree[size]) for size in sorted(max_degree.keys())]

average_random = [np.mean(random[size]) for size in sorted(random.keys())]
std_random = [np.std(random[size]) for size in sorted(random.keys())]

# Use pandas to apply a moving average
window_size = 5  # Adjust the window size as needed
df_bc = pd.DataFrame({'size': sorted(bc.keys()), 'average_bc': average_bc, 'std_bc': std_bc})
df_simple = pd.DataFrame({'size': sorted(simple.keys()), 'average_simple': average_simple, 'std_simple': std_simple})
df_pairwise = pd.DataFrame({'size': sorted(pairwise.keys()), 'average_pairwise': average_pairwise, 'std_pairwise': std_pairwise})
df_smart = pd.DataFrame({'size': sorted(smart.keys()), 'average_smart': average_smart, 'std_smart': std_smart})
df_min_degree = pd.DataFrame({'size': sorted(min_degree.keys()), 'average_min_degree': average_min_degree, 'std_min_degree': std_min_degree})
df_max_degree = pd.DataFrame({'size': sorted(max_degree.keys()), 'average_max_degree': average_max_degree, 'std_max_degree': std_max_degree})
df_random = pd.DataFrame({'size': sorted(random.keys()), 'average_random': average_random, 'std_random': std_random})

df_bc['smooth_bc'] = df_bc['average_bc'].rolling(window=window_size).mean()
df_simple['smooth_simple'] = df_simple['average_simple'].rolling(window=window_size).mean()
df_pairwise['smooth_pairwise'] = df_pairwise['average_pairwise'].rolling(window=window_size).mean()
df_smart['smooth_smart'] = df_smart['average_smart'].rolling(window=window_size).mean()
df_min_degree['smooth_min_degree'] = df_min_degree['average_min_degree'].rolling(window=window_size).mean()
df_max_degree['smooth_max_degree'] = df_max_degree['average_max_degree'].rolling(window=window_size).mean()
df_random['smooth_random'] = df_random['average_random'].rolling(window=window_size).mean()

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(df_bc['size'], df_bc['smooth_bc'], color='b', label='Binary Class')
plt.plot(df_simple['size'], df_simple['smooth_simple'], color='r', label='SimpleRank')
plt.plot(df_pairwise['size'], df_pairwise['smooth_pairwise'], color='g', label='PairwiseRank*')
plt.plot(df_smart['size'], df_smart['smooth_smart'], color='y', label='Smart')
plt.plot(df_min_degree['size'], df_min_degree['smooth_min_degree'], color='c', label='Min Degree (DBK)')
plt.plot(df_max_degree['size'], df_max_degree['smooth_max_degree'], color='m', label='Max Degree')
plt.plot(df_random['size'], df_random['smooth_random'], color='k', label='Random')

# Plot standard deviation as a dotted line
plt.fill_between(df_bc['size'], df_bc['smooth_bc'] - df_bc['std_bc'], df_bc['smooth_bc'] + df_bc['std_bc'], alpha=0.2, color='b')
plt.fill_between(df_simple['size'], df_simple['smooth_simple'] - df_simple['std_simple'], df_simple['smooth_simple'] + df_simple['std_simple'], alpha=0.2, color='r')
plt.fill_between(df_pairwise['size'], df_pairwise['smooth_pairwise'] - df_pairwise['std_pairwise'], df_pairwise['smooth_pairwise'] + df_pairwise['std_pairwise'], alpha=0.2, color='g')
plt.fill_between(df_smart['size'], df_smart['smooth_smart'] - df_smart['std_smart'], df_smart['smooth_smart'] + df_smart['std_smart'], alpha=0.2, color='y')
plt.fill_between(df_min_degree['size'], df_min_degree['smooth_min_degree'] - df_min_degree['std_min_degree'], df_min_degree['smooth_min_degree'] + df_min_degree['std_min_degree'], alpha=0.2, color='c')
plt.fill_between(df_max_degree['size'], df_max_degree['smooth_max_degree'] - df_max_degree['std_max_degree'], df_max_degree['smooth_max_degree'] + df_max_degree['std_max_degree'], alpha=0.2, color='m')
plt.fill_between(df_random['size'], df_random['smooth_random'] - df_random['std_random'], df_random['smooth_random'] + df_random['std_random'], alpha=0.2, color='k')

plt.xlabel('Graph Size')
plt.ylabel('Iterations')
plt.title('Number of Iterations Varying by Graph Size (Smoothed with Standard Deviation)')
plt.legend()
plt.grid(True)
plt.show()
```
